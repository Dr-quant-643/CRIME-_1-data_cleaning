DATA CLEANING 

```python
# CONVERSIONS 

df['DATE OCC'  ]= pd.to_datetime(df['DATE OCC'], errors='coerce')
df['Date Rptd']= pd.to_datetime(df['Date Rptd'], errors='coerce')

df[['DATE OCC', 'Date Rptd']].dtypes
```

```python
df['Hour']=df['TIME OCC']//100
df[['TIME OCC','Hour']].head()
```

```python
df.describe()
```

```python
# missing values per column with percentages
missing_values = df.isnull().sum()
missing_percentages = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentages})
print(missing_df)
```

```python
#filling all missing values with unknown # will decode later at ML 
 df_filled = df.fillna('Unknown')
df_filled.isnull().sum()
```

```python
# checking for duplicates
duplicates = df.duplicated().sum()
print(f'Number of duplicate rows: {duplicates}')
```

```python
# checking for unique values 
df.nunique()
```

```python
# getting negative values in Vict Age 
negative_ages = df[df['Vict Age'] < 0]
print(negative_ages)

# filling negative ages with zero
df['Vict Age'] = df['Vict Age'].apply(lambda x: 0 if x < 0 else x)

# confirming that there are no more negative ages
print(df[df['Vict Age'] < 0])

```

```python
#getting  unique values in Vict Sex
print(df['Vict Sex'].unique())

# if unique values not  same as  ['M', 'F'] fill with unknown
if set(df['Vict Sex'].unique()) != {'M', 'F'}:
    df['Vict Sex'] = df['Vict Sex'].fillna('Unknown') 

  # confirming the fill
    print(df['Vict Sex'].unique())
```
